```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import pandas as pd
```

```python
ds1 = tfds.load('caltech_birds2010', split='train', shuffle_files=True, as_supervised = True)
assert isinstance(ds1, tf.data.Dataset)
print(ds1)
```

```python
(ds1_train, ds1_test), info = tfds.load('caltech_birds2010', split=['train','test'], shuffle_files=True, as_supervised = True, with_info=True)
```

```python
ds2 = tfds.load('caltech_birds2011', split='train', shuffle_files=True, as_supervised = True)
assert isinstance(ds2, tf.data.Dataset)
print(ds2)
```

```python
(ds2_train, ds2_test), info = tfds.load('caltech_birds2011', split=['train','test'], shuffle_files=True, as_supervised = True, with_info=True)
```

```python
df_train1 = tfds.as_dataframe(ds1_train, info)
df_valid1 = tfds.as_dataframe(ds1_test, info)
```

```python
df_train2 = tfds.as_dataframe(ds2_train, info)
df_valid2 = tfds.as_dataframe(ds2_test, info)
```

```python
df_train = pd.concat([df_train1, df_train2])
```

```python
df_valid = pd.concat([df_valid1, df_valid2])
```

```python
y_train = df_train['label']
y_valid = df_valid['label']
```
